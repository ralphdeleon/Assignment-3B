````python
rec_week_data.loc[:,"birthmonth"].replace("^Jan","1",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Feb","2",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Mar","3",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Apr","4",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^May","5",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Jun","6",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Jul","7",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Aug","8",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Sep","9",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Oct","10",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Nov","11",regex=True,inplace=True)
rec_week_data.loc[:,"birthmonth"].replace("^Dec","12",regex=True,inplace=True)
````

````python
rec_week_data.loc[:,"birthyear"]=rec_week_data.loc[:,"birthyear"].astype(str).replace("1997","97", regex=True)
rec_week_data.loc[:,"birthyear"]=rec_week_data.loc[:,"birthyear"].astype(str).replace("1998","98", regex=True)
rec_week_data.loc[:,"birthyear"]=rec_week_data.loc[:,"birthyear"].astype(str).replace("1999","99", regex=True)
rec_week_data.loc[:,"birthyear"]=rec_week_data.loc[:,"birthyear"].astype(str).replace("2000","00", regex=True)
rec_week_data.loc[:,"birthyear"]=rec_week_data.loc[:,"birthyear"].astype(str).replace("2001","01", regex=True)


````

````python
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^Management Engineering", "ME", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^Management", "MGT", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^Information Technology Entrepreneurship", "ITE", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^Communications Technology", "BS CTM", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^Legal Management", "LM", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^BS ME", "ME", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^BS MGT", "MGT", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^BS ITE", "ITE", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^BS CTM", "CTM", regex=True)
rec_week_data.loc[:,"course"]=rec_week_data.loc[:,"course"].replace("^BS LM", "LM", regex=True)
````

````python
rec_week_data.loc[:,"phonenumber"].replace("^\+63", " ", regex=True, inplace=True)
````
